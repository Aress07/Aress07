```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("heart.csv")
df.head()
```

```python
print(df.info())
print(df.describe())
```

```python
df.duplicated().sum()
```

```python
df.dtypes
```

```python
df["Sex"].unique()
```

```python
df["Sex"] = df["Sex"].map({"M": 1, "F": 0})
df.head()
```

```python
df["ChestPainType"].unique()
```

```python
df = pd.get_dummies(df, columns=["ChestPainType"])
df.head()
```

```python
df["ChestPainType_TA"] = df["ChestPainType_TA"].astype(int)
df["ChestPainType_NAP"] = df["ChestPainType_NAP"].astype(int)
df["ChestPainType_ASY"] = df["ChestPainType_ASY"].astype(int)
df["ChestPainType_ATA"] = df["ChestPainType_ATA"].astype(int)
df.head()
```

```python
df["RestingECG"].unique()
```

```python
df = pd.get_dummies(df, columns=["RestingECG"])
```

```python
df.head()
```

```python
df["RestingECG_LVH"] = df["RestingECG_LVH"].astype(int)
df["RestingECG_Normal"] = df["RestingECG_Normal"].astype(int)
df["RestingECG_ST"] = df["RestingECG_ST"].astype(int)
df.head()
```

```python
df.dtypes
```

```python
df["ExerciseAngina"].unique()
```

```python
df["ExerciseAngina"] = df["ExerciseAngina"].map({"N": 0, "Y": 1})
df.head()
```

```python
df["ST_Slope"].unique()
```

```python
df = pd.get_dummies(df, columns=["ST_Slope"])
df.head()
```

```python
df["ST_Slope_Down"] = df["ST_Slope_Down"].astype(int)
df["ST_Slope_Flat"] = df["ST_Slope_Flat"].astype(int)
df["ST_Slope_Up"] = df["ST_Slope_Up"].astype(int)
df.head()
```

```python
df.dtypes
```

```python
x = df.drop('HeartDisease', axis=1)
y = df['HeartDisease']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
model = LogisticRegression(max_iter=10000)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy_percentage = accuracy * 100

print(f"Accuracy: {accuracy_percentage:.2f}%")
```

```python
print(classification_report(y_test, y_pred))
```

# Understanding the classification report for my heart disease model
### Here’s what the main metrics mean and how my model performed:

Support is how many actual samples there are for each class in the test data:

- Class 0 (no heart disease): 77 samples

- Class 1 (heart disease): 107 samples

Precision tells me, out of all the times the model predicted a class, how often it was correct:

- For class 0, the model was right 82% of the time it predicted “no disease.”

- For class 1, it was right 90% of the time it predicted “heart disease.”

Recall tells me, out of all actual cases of a class, how many the model correctly found:

- For class 0, the model correctly identified 87% of all “no disease” cases.

- For class 1, it found 86% of all “heart disease” cases.

F1-score is a balance between precision and recall (higher is better):

- 0.84 for class 0

- 0.88 for class 1

Accuracy is the overall percentage of correct predictions: about 86%.

Macro average is the simple average of the precision, recall, and F1-score for both classes — treating them equally.

Weighted average takes the number of samples per class into account.

Summary:
- The model is slightly better at predicting heart disease (class 1), which is great because catching those cases is important. Recall values show the model doesn’t miss too many cases in either class. Overall, the model performs well!
